```python
data1=dat.iloc[:,1:15]
```

```python
data1.head(3)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmns=KMeans(n_clusters=3,n_init=42)
```

```python
model=kmns.fit(data1)
```

```python
model.cluster_centers_ #use to find the cluster centriods
```

```python
data1['cluster']=kmns.labels_ #use to assign which datapoint  is falling to which cluster
```

```python
data1.head(4)
```

```python
print(silhouette_score(X=data1,labels=kmns.labels_))
```

```python
for k in range(2,10):
    kmns=KMeans(n_clusters=k,n_init=42)
    model=kmns.fit(data1)
    print(silhouette_score(X=data1,labels=kmns.labels_))
```

```python
d=kmns.cluster_centers_ #putting it in a dataframe
```

```python
d
```

```python
df_cluster=pd.DataFrame(d) #putting it in a dataframe
```

```python
df_cluster
```

```python
df_cluster.columns=data1.columns
```

```python
df_cluster
```

```python
#This process I am doing it for normalisation
```

```python
df_col_mean=df_cluster.mean() #finding the column means
```

```python
df_col_mean
```

```python
df_row_mean=df_cluster.transpose().mean() #finding the row means 
```

```python
df_row_mean
```

```python
df_transpose=df_cluster.transpose() #putting the transpose of the dataframe
```

```python
df_transpose
```

```python
data_f=pd.concat([df_transpose,df_col_mean],axis=1) #concatinating the the overall column means
```

```python
data_f
```

```python
data_f.columns=['cluster1','cluster2','cluster3','Overall']
```

```python
data_f
```

```python
data_f['mean1']=data_f['cluster1']-data_f['Overall']
data_f['mean2']=data_f['cluster2']-data_f['Overall']
data_f['mean3']=data_f['cluster3']-data_f['Overall']  #substracting the row means and column means from the original data
```

```python
data_f
```
